# Data Cleaning

In this section, you will need to read the CSV files into DataFrames and perform any necessary data cleaning steps. After cleaning, combine all DataFrames into a single DataFrame.

Files:

* `whale_returns.csv`: Contains returns of some famous "whale" investors' portfolios.

* `algo_returns.csv`: Contains returns from the in-house trading algorithms from Harold's company.

* `sp500_history.csv`: Contains historical closing prices of the S&P 500 Index.

## Whale Returns

Read the Whale Portfolio daily returns and clean the data

```python
# Reading whale returns
whale_csv_path = Path("Resources/whale_returns.csv")
whale_df = pd.read_csv(whale_csv_path, index_col="Date", parse_dates=True, infer_datetime_format=True)
whale_df.head()
```

```python
# Count nulls
whale_df.isnull().sum()
```

```python
# Drop nulls
#whale_no_null = whale_df.dropna()
whale_df.dropna(inplace=True)
whale_df.isnull().sum()   # make sure its cleaned
```

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data

```python
# Reading algorithmic returns
algo_csv_path = Path("Resources/algo_returns.csv")
algo_df = pd.read_csv(algo_csv_path, index_col="Date", parse_dates=True, infer_datetime_format=True)
algo_df.head()
```

```python
# Count nulls
algo_df.isnull().sum()
```

```python
# Drop nulls
algo_df.dropna(inplace=True)
algo_df.isnull().sum()           #  check data is cleaned
```

## S&P 500 Returns

Read the S&P 500 historic closing prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P 500 Closing Prices
sp_csv_path = Path("Resources/sp500_history.csv")
sp_df_close = pd.read_csv(sp_csv_path, index_col="Date", parse_dates=True, infer_datetime_format=True)
sp_df_close.sort_index(inplace=True)
sp_df_close.head()
```

```python
# Check Data Types
sp_df_close.dtypes
```

```python
#sp_df.dropna(inplace=True)
sp_df_close.isnull().sum()   # check S&P nulls too
```

```python
# Fix Data Types      # we  find that Close is string type, so we need to remover the $ sign
sp_df_close["Close"] = sp_df_close["Close"].str.replace("$", "")
print(sp_df_close.head())
sp_df_close["Close"] = sp_df_close["Close"].astype("float")
sp_df_close.dtypes
```

```python
# Calculate Daily Returns
sp_daily_returns = sp_df_close.pct_change()
sp_daily_returns.head()
```

```python
# Drop nulls
print(sp_daily_returns.isnull().sum())
sp_daily_returns.dropna(inplace=True)
print(sp_daily_returns.isnull().sum())     # check nulls are dropped
sp_daily_returns.head()
```

```python
# Rename `Close` Column to be specific to this portfolio.
#sp_daily_returns.columns
sp_daily_returns.rename(columns={"Close": "SPClose"},  inplace=True)
```

```python
# whale returns - whale is already returns
#whale_returns  = whale_df.pct_change().dropna()
#whale_returns.head()
```

```python
# Algo returns - already returns in the file. no need of pct-change
#algo_returns = algo_df.pct_change().dropna()
#algo_returns.sample(10)
```

## Combine Whale, Algorithmic, and S&P 500 Returns

```python
# Join Whale Returns, Algorithmic Returns, and the S&P 500 Returns into a single DataFrame with columns for each portfolio's returns.
combined_df = pd.concat([whale_df, algo_df, sp_daily_returns], axis="columns", join="inner")
combined_df.head()
```

### Calculate and Plot Beta for a chosen portfolio and the S&P 500

```python
# Calculate covariance of a single portfolio
#print(combined_df.info())
#combined_df['BERKSHIRE HATHAWAY INC']
covariance_berkshire = combined_df['BERKSHIRE HATHAWAY INC'].rolling(window=50).cov(combined_df['SPClose'])
#print(covariance_berkshire)
#covariance_berkshire.plot()
# Calculate variance of S&P 500
variance_SP = combined_df['SPClose'].rolling(window=50).var()
#print(variance_SP)
# Computing beta
berkshire_beta = covariance_berkshire / variance_SP
#print(berkshire_beta)
# Plot beta trend
berkshire_beta.plot(figsize=(20, 10), title='Berkshire Beta w S&P')
```

# Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. After all, if you could invest in one of two portfolios, and each offered the same 10% return, yet one offered lower risk, you'd take that one, right?

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Annualized Sharpe Ratios
sharpe_ratios = (combined_df.mean() * 252) / (combined_df.std() * np.sqrt(252))
print(sharpe_ratios)
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe_ratios.plot(kind="bar", title="Sharpe Ratios")
```

For this demo solution, we fetch data from three companies listes in the S&P 500 index.

* `GOOG` - [Google, LLC](https://en.wikipedia.org/wiki/Google)

* `AAPL` - [Apple Inc.](https://en.wikipedia.org/wiki/Apple_Inc.)

* `COST` - [Costco Wholesale Corporation](https://en.wikipedia.org/wiki/Costco)

* `NVDA` - [Nvidia Corporation](https://en.wikipedia.org/wiki/Nvidia)

```python
# Reading data from 1st stock TESLA - end date 5/7/21
tsla_csv_path = Path("Resources/TSLA.csv")
tsla_df = pd.read_csv(tsla_csv_path, index_col="Date", parse_dates=True, infer_datetime_format=True)
tsla_df.isna().sum()
tsla_df.dropna(inplace=True)
tsla_df.isna().sum()
tsla_df.rename(columns = {'Close': 'TSLA'}, inplace = True)
tsla_df.tail()
```

```python
# Reading data from 2nd stock - Google
goog_csv_path = Path("Resources/GOOG.csv")
goog_df = pd.read_csv(goog_csv_path, index_col="Date", parse_dates=True, infer_datetime_format=True)
goog_df.isna().sum()
goog_df.dropna(inplace=True)
goog_df.rename(columns = {'Close': 'GOOG'}, inplace = True)
goog_df.tail()
```

```python
# Reading data from 3rd stock - APple
aapl_csv_path = Path("Resources/AAPL.csv")
aapl_df = pd.read_csv(aapl_csv_path, index_col="Date", parse_dates=True, infer_datetime_format=True)
aapl_df.isna().sum()
aapl_df.dropna(inplace=True)
aapl_df.rename(columns = {'Close': 'AAPL'}, inplace = True)
aapl_df.tail()
```

```python
# Reading data from 4th stock - Invitae Corp
nvta_csv_path = Path("Resources/NVTA.csv")
nvta_df = pd.read_csv(nvta_csv_path, index_col="Date", parse_dates=True, infer_datetime_format=True)
nvta_df.isna().sum()
nvta_df.dropna(inplace=True)
nvta_df.rename(columns = {'Close': 'NVTA'}, inplace = True)
nvta_df.tail()
```

```python
# Reading data from 5th stock - CRISPR Therapeutics
crsp_csv_path = Path("Resources/CRSP.csv")
crsp_df = pd.read_csv(crsp_csv_path, index_col="Date", parse_dates=True, infer_datetime_format=True)
crsp_df.isna().sum()
crsp_df.dropna(inplace=True)
crsp_df.rename(columns = {'Close': 'CRSP'}, inplace = True)
crsp_df.tail()
```

```python
# Read from S&P latest data
spy_csv_path = Path("Resources/SPY.csv")
spy_df = pd.read_csv(spy_csv_path, index_col="Date", parse_dates=True, infer_datetime_format=True)
spy_df.isna().sum()
spy_df.dropna(inplace=True)
spy_df.rename(columns = {'Close': 'SPY'}, inplace = True)
spy_df.tail()
```

```python
# Combine all stocks in a single DataFrame
custom_combined_df = pd.concat([tsla_df, goog_df, aapl_df,  nvta_df, crsp_df, spy_df], axis="columns", join="inner")
custom_combined_df.sort_index      # make sure dates are sorted
custom_combined_df.tail()
```

```python
# Reset Date index
custom_combined_df.index = pd.DatetimeIndex(custom_combined_df.index).normalize()
# custom_combined_df.reset_index(inplace=True)
# type(custom_combined_df.index)

# custom_combined_df.set_index('index')
print(custom_combined_df.tail())
```

```python
# Reorganize portfolio data by having a column per symbol
```

```python
# Calculate daily returns
custom_port_returns_df = custom_combined_df.pct_change()
# Drop NAs
custom_port_returns_df.dropna(inplace=True)
# Display sample data
custom_port_returns_df.head()
```

## Calculate the weighted returns for the portfolio assuming an equal number of shares for each stock

```python
# Set weights
weights = [1/6, 1/6, 1/6, 1/6, 1/6, 1/6]

# Calculate portfolio return
custom_port_returns_weighted_df = custom_port_returns_df.dot(weights)
# Display sample data
custom_port_returns_weighted_df.head()
```

## Join your portfolio returns to the DataFrame that contains all of the portfolio returns

```python
# Join your returns DataFrame to the original returns DataFrame
all_df = pd.concat([custom_port_returns_df, combined_df], axis="columns", join="inner")
all_df.head()
```

```python
# Only compare dates where return data exists for all the stocks (drop NaNs)
all_df.isna().sum()    # no nulls exist in all combined df already
all_df.dropna(inplace=True)
print(all_df.isna().sum())
```

```python
# Calculate and plot the correlation
all_df_correlation = all_df.corr()
#print(all_df_correlation)
sns.heatmap(all_df_correlation, vmin=-1, vmax=1)
```

### Calculate and Plot Rolling 60-day Beta for Your Portfolio compared to the S&P 500

```python
# Calculate and plot Beta
covariance_nvta = all_df['NVTA'].rolling(window=60).cov(all_df['SPY'])
covariance_crsp = all_df['CRSP'].rolling(window=60).cov(all_df['SPY'])
covariance_goog = all_df['GOOG'].rolling(window=60).cov(all_df['SPY'])
covariance_tsla = all_df['TSLA'].rolling(window=60).cov(all_df['SPY'])
covariance_aapl = all_df['AAPL'].rolling(window=60).cov(all_df['SPY'])
#print(covariance_berkshire)
#covariance_berkshire.plot()
# Calculate variance of S&P 500
variance_SPY = all_df['SPY'].rolling(window=60).var()
#print(variance_SP)
# Computing beta
nvta_beta = covariance_nvta / variance_SPY
crsp_beta = covariance_crsp / variance_SPY
tsla_beta = covariance_tsla / variance_SPY
goog_beta = covariance_goog / variance_SPY
aapl_beta = covariance_aapl / variance_SPY
#print(berkshire_beta)
# Plot beta trend
ax = nvta_beta.plot(figsize=(20, 10), legend=True, label = 'NVTA')
crsp_beta.plot(figsize=(20, 10), ax=ax, legend=True, label = 'CRSP')
tsla_beta.plot(figsize=(20, 10), ax=ax, legend=True, label ='TSLA')
goog_beta.plot(figsize=(20, 10), ax=ax, legend=True, label ='GOOG')
aapl_beta.plot(figsize=(20, 10), ax=ax, legend=True, label ='AAPL')
```
